```python
import pandas as pd
from collections import defaultdict
```

```python
data = pd.read_csv('user_comment.csv')
```

```python
data
```

```python
data.head()
```

```python
names = data['author_name']
```

```python
d = defaultdict(lambda: len(d))
alist = [d[x]+1 for x in names]
```

```python
author_id = pd.DataFrame(alist)
```

```python
data['author_id'] = author_id
```

```python
alist=[]
for name in names:
    email = str(name) + '@bestchoice.com'
    alist.append(email)
```

```python
emails = pd.DataFrame(alist)
```

```python
emails
```

```python
data['email'] = emails
```

```python
data
```

```python
data.to_csv('user.csv')
```
